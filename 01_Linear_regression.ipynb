```python
from sklearn.linear_model import LinearRegression
import pandas as pd
data={
    'Area':[2600,3000,3200,3600,4000],
    'Price':[550000,565000,610000,680000,725000]
}
df=pd.DataFrame(data)


reg =linear_model.LinearRegression() # create a object
reg.fit(df[['Area']],df.Price) # train the model

#predict our model
reg.predict(pd.DataFrame([[2000]], columns=['Area']))


#array([452191.78082192]) predict the price regarding area
reg.coef_
reg.intercept_

Y = reg.coef_ * 2000 + reg.intercept_
Y

'''
Linear reg: Y=mX + c
135.78767123 * 2000 +180616.43835616432
Y = 452191.78081616433
'''
```

```python
new=pd.DataFrame({
    'Area':[1000,1500,2300,3600,4800,5600,4300,5100]
})
reg.predict(new)
new
```
